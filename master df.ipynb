```python
import numpy as np
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
```

```python
file = 'Master Dataset 2020.xlsx'
colnames = ['Record_ID','Policy_ID','Driver_ID','Vehicle_ID','Vehicle_Model_Year','Location_ID','Total_Premium']
risk_df= pd.read_excel(file, sheet_name ='Risk_Fact ', skiprows=0, names=colnames)
risk_df.head()
risk_df.to_pickle('risk.p')
```

```python
colnames=['Vehicle_ID','Car_Make','Car_Model','Average_Price']
vehicle_df = pd.read_excel(file, sheet_name='Vehicles_Dimension', skiprows=0, names=colnames)
vehicle_df.head()
vehicle_df.to_pickle('vehicle.p')
```

```python
master_df=pd.merge(risk_df,vehicle_df, on='Vehicle_ID',how='left')
master_df.head()
```

```python
colnames=['Policy_ID','Policy Effective Date','Policy Expiration Date','Agency ID','Underwriting Company ID','Cancel Date']
policy_df = pd.read_excel(file, sheet_name='Policy_Dimension', skiprows=0, names=colnames)
policy_df.head()
policy_df.to_pickle('policy.p')
```

```python
master_df2=pd.merge(master_df,policy_df, on='Policy_ID',how='left')
master_df2.head()
```

```python
colnames=['Agency ID','Agency Name','Agency City','Agency State']
agency_df = pd.read_excel(file, sheet_name='Agency_Dimension', skiprows=0, names=colnames)
agency_df.head()
agency_df.to_pickle('agency.p')
```

```python
master_df3=pd.merge(master_df2,agency_df, on='Agency ID',how='left')
master_df3.head()
```

```python
colnames=['Location_ID','ZIP','Primary_City','State','County','Estimated_population_2013']
location_df = pd.read_excel(file, sheet_name='Location_Dimension', skiprows=0, names=colnames)
location_df.head()
location_df.to_pickle('location.p')
```

```python
master_df4=pd.merge(master_df3,location_df, on='Location_ID',how='left')
master_df4.head()
```

```python
master_df5=pd.merge(master_df4, population_df, on=['ZIP','Estimated_population_2013'], how='left')
master_df5.head()
```

```python
colnames=['Driver_ID','First Name','Last Name','Number of Violations','Number of Accidents','Marital Status','Gender','Miles to work','Primary Vehicle Use','Date of Birth']
driver_df = pd.read_excel(file, sheet_name=' Driver_Dimension', skiprows=0, names=colnames)
driver_df.head()
driver_df.to_pickle('driver.p')
```
